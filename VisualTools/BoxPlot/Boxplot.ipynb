```python
import seaborn as sns
import pandas as pd
import matplotlib.pyplot as plt

sns.set_theme(style="whitegrid")
dataframe = pd.read_csv('data.csv', encoding="utf-8")
dataframe
```

```python
sns.set_style('whitegrid', {'grid.linestyle': '--'})
sns.set_context({'figure.figsize':[23, 6]})
sns.set(context='notebook', style='ticks', font_scale=1.6)

# plt.grid()  #添加网格gird, 添加在这里只有横着的网格

ax = sns.boxplot(x="anatomical structure", y="Dice", hue="methods",
                 data=dataframe, palette="Paired", linewidth=0.8, width=0.7, fliersize=1)

# sns.move_legend(ax, "upper left", bbox_to_anchor=(1, 1))
sns.move_legend(
    ax, "lower center",
    bbox_to_anchor=(.5, 1), ncol=7, title=None, frameon=False
)

ax.set_xticklabels(ax.get_xticklabels(), rotation=75)
plt.grid()  #添加网格gird, 添加在这里横竖都有

scatter_fig = ax.get_figure()
scatter_fig.savefig('./boxplot.svg', dpi = 800, bbox_inches='tight')
scatter_fig.savefig('./boxplot.pdf', dpi = 800, bbox_inches='tight')
scatter_fig.savefig('./boxplot.eps', dpi = 800, bbox_inches='tight')

# plt.show()
```
